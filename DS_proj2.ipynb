# Read Table salaries using SQL 

```python
import numpy as np
salaries= pd.read_sql_query("SELECT * from salaries", con)
steam = list(salaries.teamID)

syears = list(salaries.yearID)
team = list(teams.teamID)
years = list(teams.yearID)

salary = list(salaries.salary)

tpayY=dict()
cnt=dict()
winp=dict()
winn=dict()
payTeam=dict()#part 2
for i in range(len(team)):
    winn[team[i]]=dict()
    payTeam[team[i]]=dict()
for i in range(len(winn)):
    for j in range(len(years)):
        winn[team[i]][years[j]]=0
        payTeam[team[i]][years[j]]=0
for i in range(len(winn)):
    for j in range(len(years)):
        payTeam[team[i]][years[j]]+=salary[j]    
        
for i in range(len(years)):
    tpayY[years[i]]= dict()
    winp[years[i]]= dict()
    
for i in range(len(years)):
    tpayY[years[i]][team[i]]=0
    winp[years[i]][team[i]]=0
    cnt[years[i]]=0
for i in range(len(syears)):
    try:
            tpayY[syears[i]][steam[i]]+=salary[i]
    except:
        continue
for i in range(len(years)):
    cnt[years[i]]=cnt[years[i]]+1

```

```python
teams["payroll"]=0
teams["winp"]=0
for i, row in teams.iterrows():
    teams.at[i,'winp']=(teams.at[i,'W']/teams.at[i,'G'])*100

    try:
        winn[teams.at[i,'teamID']][teams.at[i,'yearID']]=teams.at[i,'winp']
        winp[teams.at[i,'yearID']][teams.at[i,'teamID']]= teams.at[i,'winp']
        teams.at[i,'payroll']=tpayY[teams.at[i,'yearID']][teams.at[i,'teamID']] 
    except:
        continue

```

# including other columns:
we should here include some columns like total payroll ,average payroll and the standard deviation.

```python
teamid=set(teams.teamID)
yearid=set(teams.yearID)
teamid=list(teamid)
yearid=list(yearid)
tpayT=dict()
avg_pay=dict()
spayY=dict()
year_payrolls = dict()


for i, row in teams.iterrows():
    tpayY[teams.at[i,'yearID']][teams.at[i,'teamID']]=teams.at[i,'payroll'] #update payroll of each team per year
for j in range(len(teamid)):
    tpayT[teamid[j]]=0
for i in range(len(yearid)):
    avg_pay[yearid[i]]=0 
    spayY[yearid[i]]=0
    
for i in range(len(tpayT)):
    for j in range(len(teamid)):
        try:
            tpayT[teamid[j]]+=tpayY[yearid[i]][teamid[j]]#get total payroll for each team
            avg_pay[yearid[i]]+=tpayY[yearid[i]][teamid[j]] #get total payroll per team  
        except:
            continue
for i in range(len(avg_pay)):
    avg_pay[yearid[i]]=avg_pay[yearid[i]]/cnt[yearid[i]] ##get averge of payroll in year
    
for i in tpayY:
    Map = tpayY[i]
    payrolls = []
    for j in Map:
        payrolls.append(Map[j])
    year_payrolls[i]=payrolls
for i,j in zip(year_payrolls,spayY):
    spayY[j]=np.std(year_payrolls[i])
```

```python
teams["TotalPayroll"]=0
teams["avgPayroll"]=0
teams["stander_payroll"]=0
ploteam=dict()
yearab=list()
teamid=list(teams.teamID)
yearid=list(teams.yearID)
for j in range(len(yearid)):
    if yearid[j]>=1990:
        yearab.append(yearid[j])
yearab=set(yearab)
yearab=list(yearab)
for j in range(len(teamid)):
    ploteam[teamid[j]]=dict()
for j in range(len(yearab)):
    ploteam[teamid[j]][yearab[j]]=0

for i, row in teams.iterrows():
    try:
        if teams.at[i,'yearID']>=1990:
            ploteam[teams.at[i,'teamID']][teams.at[i,'yearID']]=teams.at[i,'payroll']
        teams.at[i,'avgPayroll']=avg_pay[teams.at[i,'yearID']]
        teams.at[i,'TotalPayroll']=tpayT[teams.at[i,'teamID']]
        teams.at[i,"stander_payroll"]=spayY[teams.at[i,'yearID']]
    except:
        continue
```

# Correlation between payroll and winning percentage

```python
import matplotlib.pyplot as plt
f1=plt.figure()
f2=plt.figure()
ax1=f1.add_subplot(111)
ax2=f2.add_subplot(111)
teamsname=teams.teamID.unique()
numPieriods=5
for i in range(3,30):
    name=teamsname[i]
    data=teams.query('teamID==\''+name+'\'')
    length=data.teamID.count()//numPieriods
    if length ==0:
        continue
    meanP=[]
    meanW=[]
    payroll_arr=np.array(data.payroll)
    win_arr=np.array(data.winp)
    for j in range(numPieriods):
        meanP.append(np.mean(payroll_arr[j*length:(i+1)*length]))
        meanW.append(np.mean(win_arr[j*length:(i+1)*length]))
    ax1.scatter(np.linspace(1990,2014,num=numPieriods),meanP,label=name)
  
    ax2.scatter(np.linspace(1990,2014,num=numPieriods),meanW,label=name)
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper left')
plt.show()
   
    
    
```

As we can see from the graph, the average payroll (central tendency) of baseball teams in a given year has almost always increased as time goes on. On average, a team payroll is more in an given year than from the year before.
